```python
df.sort_values("Year") #gives years in sorting values from 2016 to 2023
```

```python
df.sort_values("Year", ascending = False) #gives descending order from 2023 to 2016
```

```python
#sorting with multiple columns 
df.sort_values(["Year" ,"IMDb"])# when year has tie then do sorting on base of IMDb
```

```python
df2 = df.sort_values(["Year" ,"IMDb"]).copy()
df2
```

```python
df2.sort_index() # it will sort the index in order
```

```python
df2.reset_index() # it will gives the new index in order and it will not change the df2 data frame it a view not a copy
```

```python
df2 # hence proved our original data frame is not changed it will change it we use .copy method
```

when reset the dataframe df2 it gives original index as well the new index in order but if want remove the old index and just want new index in order then use drop=True inside reset_index

```python
df2.reset_index(drop=True, inplace=True) #its give view not copy if want copy without copy method use inplace=True
```

```python
df2#Now df2 is updated without using .copy by using inplace=True
```

```python
#Ranking
df2['Rank']=df2['IMDb'].rank() # add a rank column with ranking according to IMDb
df2
```

```python
df2['Rank']=df2['IMDb'].rank(ascending = False) #ranking descending order
df2
```

```python
df2['Rank']=df2['IMDb'].rank(ascending = False, method="dense")#make the ranking in integers and same imdb share same ranking
df2
```

```python
#problem 1: Sort the DataFrame by the Score column in ascending order.
import pandas as pd
data = {
    'Name': ['Ali', 'Zara', 'Hina', 'Fahad'],
    'Score': [88, 75, 95, 60]
}
df = pd.DataFrame(data)
df
```

```python
df=df.sort_values(["Score"])
df
```

```python
#Problem 2: Sort by Multiple Columns... Sort by Name ascending and Score descending within each name group.
data = {
    'Name': ['Ali', 'Zara', 'Ali', 'Zara'],
    'Subject': ['Math', 'Math', 'English', 'English'],
    'Score': [80, 95, 85, 90]
}
df = pd.DataFrame(data)
df
```

```python
df.sort_values(['Name' ,'Score'], ascending=[True, False])
```

```python
#problem 3: Create a new column "Rank" that ranks students based on Score (highest = rank 1). Use dense ranking (no gaps in ranks).
data = {
    'Student': ['Adeel', 'Sana', 'Hamza', 'Zainab'],
    'Score': [70, 90, 90, 85]
}
df = pd.DataFrame(data)
df
```

```python
df["Rank"]=df["Score"].rank(ascending=False, method="dense")
```

```python
df
```
