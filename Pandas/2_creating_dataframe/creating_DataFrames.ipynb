Let’s look at different ways to create a Pandas DataFrame

```python
#from python list
import pandas as pd
data = [['hadia', 30], ['abuzar', 44], ['john', 88], ['jack', 28]] #it is a list inside list
data
```

```python
pd.DataFrame(data, columns=['Name', 'Marks'])
```

```python
#from python dictionary
data = {"a":[4, 6, 3], "b":[56, 34, 55]}
data
```

```python
pd.DataFrame(data)
```

```python
#from numpy arrays
import numpy as np
arr = np.array([[2, 3], [4, 6]])
arr
```

```python
df = pd.DataFrame(arr, columns=["A","B"])
```

```python
df
```

```python
# Problem 1: From Dictionary of Lists
data = {
    "Name": ["Ali", "Sara", "Omar"],
    "Age": [22, 25, 24],
    "City": ["Lahore", "Karachi", "Islamabad"]
}
pd.DataFrame(data)
```

```python
#Problem 2: From List of Dictionaries
students = [
    {"name": "Hadia", "language": "Python"},
    {"name": "Ayesha", "language": "Java"},
    {"name": "Bilal", "language": "C++"}
]
pd.DataFrame(students)
```

```python
#Problem 3: From Lists with Custom Column Names
data = [
    ["Ali", 85],
    ["Sara", 90],
    ["Omar", 78]
]
pd.DataFrame(data, columns=["Name", "Marks"])
```

```python
# Problem 4: Create DataFrame with Custom Index
data = {
    "Subject": ["Math", "Science", "English"],
    "Score": [88, 92, 85]
}
pd.DataFrame(data, index=["Test1", "Test2", "Test3"])
```
