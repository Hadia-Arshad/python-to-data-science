```python
#Problem 1: Remove Leading and Trailing Spaces
import pandas as pd
data = {
    'Name': [' Ali', 'Zara ', '  Hina  ', 'Fahad'],
    'City': ['Lahore ', ' Karachi', ' Islamabad ', 'Multan']
}
df = pd.DataFrame(data)
print(df)
```

```python
df['Name']=df['Name'].str.strip()
df['City']=df['City'].str.strip()
print(df)
```

```python
# problem 2: Fill Missing Values
#Replace NaN in Math with the mean of the column.
#Replace NaN in English with the median of the column.

data = {
    'Student': ['Adeel', 'Sana', 'Hamza', 'Zainab'],
    'Math': [88, None, 92, None],
    'English': [76, 85, None, 90]
}
df = pd.DataFrame(data)
df
```

```python
df['Math']=df['Math'].fillna(df['Math'].mean())
df
```

```python
df['English']=df['English'].fillna(df['English'].median())
df
```

```python
# problem 3: Convert the ID column to integer type and Score column to float type.
data = {
    'ID': ['001', '002', '003'],
    'Score': ['85', '90', '88']
}
df = pd.DataFrame(data)
df
```

```python
df['ID']= df['ID'].astype(int)
df
```

```python
df['Score']=df['Score'].astype(float)
df
```

```python
#problem 4: Problem 4: Replace Invalid Entries
#Replace ages < 0 or > 120 with NaN.
#Replace "Unknown" and "Other" in the Gender column with "Unspecified".
import pandas as pd 
import numpy as np
data = {
    'Age': [22, 25, -5, 30, 1000],
    'Gender': ['Male', 'Female', 'Unknown', 'Male', 'Other']
}
df = pd.DataFrame(data)
df
```

```python
df['Age'] = df['Age'].apply(lambda x: x if 0 < x < 120 else np.nan)
df
```

```python
df['Gender'] = df['Gender'].replace(['Unknown', 'Other'], 'Unspecified')
```

```python
df
```

```python
#problem 5: Drop duplicate rows. Reset the index of the DataFrame (no inplace=True).
data = {
    'Name': ['Ali', 'Zara', 'Ali', 'Hina', 'Zara'],
    'Score': [85, 90, 85, 92, 90]
}
df = pd.DataFrame(data)
df
```

```python
df = df.drop_duplicates()
df
```

```python
#now reset the index
df.reset_index(drop=True, inplace=True)
df
```
