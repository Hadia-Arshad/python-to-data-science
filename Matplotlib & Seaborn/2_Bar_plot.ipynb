Bar charts are used to compare quantities across categories. They are easy to read and powerful for visual analysis.

```python
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
#Lets compare Sachin, Sehwag, and Kohli side-by-side for the same years.
import numpy as np

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width = 0.25
#The value of width ranges from 0 to 1 (default 0.8); it can go above 1, but bars will start to overlap.
plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
```

```python
# Horizontal bar chart
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()# compact the chart
```

# Adding Value Labels with plt.text()
You can use plt.text() to display values on top of bars, making your chart easier to read. It's especially useful in presentations and reports.

```python
import matplotlib.pyplot as plt

players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color="skyblue")

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 50, str(runs[i]), ha='center')

plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
```

Problem 3:Multiple Bars (Grouped Bar Chart)
Compare the marks of two subjects (Math and Science) for 4 students:
Show them side-by-side with a legend.

```python
import matplotlib.pyplot as plt
import numpy as np
Students = ["Ali", "Sara", "Omar", "Hina"]
Math = [85, 90, 78, 92]
Science = [88, 85, 82, 95]
x=np.arange(len(Students))
width=0.25 #width of each bar
plt.bar(x-width/2, Math, width=width, label="Math", color='blue')
plt.bar(x+width/2, Science, width=width, label='Science', color='red')
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students Marks Comparison")
plt.xticks(x, Students)  # Set student names as x-axis labels
plt.legend()
plt.tight_layout()
```

problem 4: Stacked Bar Chart
Plot a stacked bar chart showing the number of apples and oranges sold in 4 months:
Use different colors for apples and oranges.

```python
import matplotlib.pyplot as plt
# Data
months = ["Jan", "Feb", "Mar", "Apr"]
apples = [50, 40, 70, 60]
oranges = [30, 50, 40, 80]
plt.bar(months, apples, label="Apples", color='red')
plt.bar(months, oranges, bottom=apples, label="Oranges", color='orange')
plt.xlabel("Months")
plt.ylabel("Fruits Sales")
plt.title("Fruits Sold in 4 Months")
plt.legend()
plt.tight_layout()
```

problem 5: Custom Bar Chart
Create a bar chart showing average monthly temperatures in Celsius for 6 months:
Change bar color to gradient shades of red.
Add value labels on top of each bar.

```python
import matplotlib.pyplot as plt
import numpy as np
Months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
Temperatures = [12, 15, 20, 25, 30, 35]
plt.bar(Months, Temperatures)
# Gradient shades of red
reds = plt.cm.Reds(np.linspace(0.4, 0.9, len(Months)))
plt.bar(Months, Temperatures, color=reds)
for i in range(len(Months)):
    plt.text(i, Temperatures[i]+0.5, str(Temperatures[i]), ha='center')

plt.xlabel('Months')
plt.ylabel('Temperature')
plt.title("Temperature per Month")
plt.tight_layout()
```

Assignment:
Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
Then create a horizontal bar chart showing total runs scored in their debut 5 years.

```python
import matplotlib.pyplot as plt
import numpy as np
# Data for 5 selected years
years = [2005, 2006, 2007, 2008, 2009]
sachin_runs = [900, 1100, 800, 1300, 1000]
kohli_runs = [0, 0, 300, 1200, 1500]
sehwag_runs = [1200, 1000, 900, 1100, 800]
x = np.arange(len(years))  # index positions
width = 0.25
#The value of width ranges from 0 to 1 (default 0.8); it can go above 1, but bars will start to overlap.
plt.figure(figsize=(8, 5))# hepls to create different fogures
plt.bar(x - width, sachin_runs, width=width, label="Sachin")
plt.bar(x, kohli_runs, width=width, label="Kohli", alpha=0.7)
plt.bar(x + width, sehwag_runs, width=width, label="sehwag", alpha=0.5)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Comparing runs over 5 years")
plt.xticks(x, years)
plt.tight_layout()
plt.legend()
plt.figure(figsize=(8, 5))
total_runs = [sum(sachin_runs), sum(kohli_runs), sum(sehwag_runs)]
players = ["Sachin", "Kohli", "Sehwag"]
plt.barh(players, total_runs, color=['blue', 'orange', 'green'])
plt.xlabel("Total Runs")
plt.title("Total Runs in Debut 5 Years")
plt.show()
```
