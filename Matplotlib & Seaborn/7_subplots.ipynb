Subplots allow you to show multiple plots in a single figure, side-by-side or in a grid layout. This is helpful when comparing different datasets or aspects of the same data.



```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]

# Create a figure with 1 row and 2 columns
plt.subplot(1, 2, 1)  # (rows, cols, plot_number)
plt.plot(x, y1)
plt.title('Double of x')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Square of x')

plt.tight_layout()
plt.show()
#plt.subplot(1, 2, 1) means 1 row, 2 columns, and we're plotting in the 1st subplot.
```

```python
#22 Grid of Subplots
# More variations of x
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]

plt.figure(figsize=(8, 6))  # Optional: make it bigger

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('x * 2')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('x squared')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('sqrt(x)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('10 - x')

plt.tight_layout()
plt.show()
#This lays out 4 plots in a 2x2 grid. plt.tight_layout() avoids overlapping titles and labels.
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))# any one or more graph from the plots 

axs[0].plot(x, y1)# comes out the 1st plot
axs[0].set_title('x * 2')#gives title to forst plot

axs[1].plot(x, y2)#comes out the 2nd plot 
axs[1].set_title('x squared')# add title on 2nd plot
 
fig.suptitle('Simple Comparison Plots', fontsize=14)# will add title on the both plots which comes out from the above plot
fig.tight_layout()
fig.subplots_adjust(top=0.85)  # So title doesn't overlap
fig.savefig('my_plots.png')    # Save as image
plt.show()
#axs is for working on individual plots
#fig is for settings that apply to the whole figure
#plt.subplots() returns a figure and a list/array of axes objects. This is more flexible and cleaner, especially for loops or advanced customizations.
```

```python
#Looping Over Subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
ys = [y1, y2, y3, y4]
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']

for i in range(2):
    for j in range(2):
        idx = i * 2 + j
        axs[i, j].plot(x, ys[idx])
        axs[i, j].set_title(titles[idx])

plt.tight_layout()
plt.show()
#This approach works well when dealing with dynamic or repetitive data series.
```

```python
# Problem 1: Create a figure with two subplots (1 row, 2 columns).
# Left plot: Line plot showing months vs average rainfall (blue line, marker='o').
# Right plot: Bar chart showing months vs number of sunny days (orange color).
# Add appropriate titles and axis labels for both plots.
import matplotlib.pyplot as plt
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
rainfall = [78, 65, 92, 110, 130, 85]
sunny_days = [5, 7, 6, 8, 9, 10]
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(months, rainfall, color='blue', marker='o')
plt.xlabel("Months")
plt.ylabel("Rain Fall")
plt.grid(True, linestyle="--", alpha=0.5)
plt.title('months vs average rainfall')
plt.subplot(1, 2, 2)
plt.bar(months, sunny_days, color='orange')
plt.xlabel("Months")
plt.ylabel("Sunny Days")
plt.title('months vs number of sunny days')
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()

```

```python
# Problem 2: Create a figure with two subplots (2 rows, 1 column).
# First plot: Scatter plot of study hours vs exam scores (red circles, size=100, alpha=0.7).
# Second plot: Histogram of exam scores (green color, bins=5).
# Add titles and axis labels for both plots.
import matplotlib.pyplot as plt
study_hours = [2, 4, 5, 6, 7, 8, 9, 10]
exam_scores = [50, 55, 60, 65, 70, 75, 85, 90]
plt.figure(figsize=(6, 6))
plt.subplot(2, 1, 1)
size=100
plt.scatter(study_hours, exam_scores, color='red', marker='o', s=size, alpha=0.7)
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours VS Exam Scores")
plt.grid(True, linestyle='--', alpha=0.5)
plt.subplot(2, 1, 2)
plt.hist(exam_scores, color='green', bins=5, edgecolor='black', alpha=0.5)
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Scores")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()


```

```python
# Problem 3: Create a figure with three subplots (1 row, 3 columns).
# Plot 1: Line plot showing days vs temperature (red line with square markers).
# Plot 2: Bar plot of days vs humidity (blue color).
# Plot 3: Scatter plot of temperature vs humidity (purple triangles, alpha=0.6).
# Add suitable titles for each subplot.
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
temperature = [22, 25, 27, 24, 23]
humidity = [60, 58, 65, 70, 68]
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
plt.plot(days, temperature, color='red', marker='s')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Days Vs Temperature")
plt.grid(True)
plt.subplot(1, 3, 2)
plt.bar(days, humidity, color='blue')
plt.title("Days Vs Humidity")
plt.xlabel("Days")
plt.ylabel("Humidity")
plt.grid(True, alpha=0.5)
plt.subplot(1, 3, 3)
plt.scatter(temperature, humidity, marker='^', color='purple', alpha=0.6)
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature Vs Humidity")
plt.grid(True)
plt.tight_layout()
```

```python
# Problem 4: Create a figure with two subplots (1 row, 2 columns).
# Left plot: Pie chart of sales distribution among 4 products.
# Right plot: Bar chart of products vs profit.
# Add titles for both plots.
import matplotlib.pyplot as plt
products = ["A", "B", "C", "D"]
sales = [200, 150, 300, 250]
profit = [50, 40, 70, 60]
plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1)
plt.pie(sales, labels=products, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title("Sales Distribution")
plt.subplot(1, 2, 2)
plt.bar(products, profit, color='orange', label='Profit')
plt.xlabel("Products")
plt.ylabel("Profit")
plt.title("Product vs Profit")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
```
