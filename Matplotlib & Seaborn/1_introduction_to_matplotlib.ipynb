```python
# Real Data Example: Sachin Tendulkar’s Runs Over Time
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

```python
# adding x and y
plt.plot(years, runs)
plt.xlabel("years")
plt.ylabel("runs scored")
plt.title("Sachin Tendulkar’s yearly runs")
```

```python
# Multiple Lines in One Plot
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
# Also label like this 
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli)
plt.plot(years, sehwag)
plt.legend(["kohli", "sehwag"]) 

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.show()
```

```python
#Format strings 
plt.plot(years, kohli, 'ro--', label = "kohli")#red circles with dash lines
plt.plot(years, sehwag, 'g^:', label = "sehwag")# green triangles dotted 
plt.legend()
```

```python
#farmat string is not much more readable so theres another way of formating
#Color and Line Style Arguments
plt.plot(years, kohli, color='orange', linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.', label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar")
plt.legend()
```

```python
# Line Width and Layout Tweaks
plt.plot(years, kohli, linewidth=3, label="Kohli") # linewidth make lines thiker
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)# add grid in back griund
plt.tight_layout()# show graph in tight layout
plt.show()
```

```python
#problem 6: You have sales data for two products
Months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ProductA = [500, 520, 510, 530, 550, 560, 570, 590, 600, 620, 610, 630]
ProductB = [450, 460, 455, 470, 480, 490, 500, 515, 525, 540, 530, 550]
```

Task:

Plot both products on the same chart.

Product A: solid red line with circle markers.

Product B: dashed blue line with square markers.

Add axis labels, title, grid lines, and a legend.

```python
plt.plot(Months, ProductA, marker='o', color='red', linestyle='-', label="ProductA")
plt.plot(Months, ProductB, marker='s', color='blue', linestyle='--', label="ProductB")
plt.xlabel("Months")
plt.ylabel("Product Sales")
plt.grid(True)
plt.legend()
plt.title("Sales Data")
```

```python
#problem 7: You have website visits for 15 days
Days = list(range(1, 16))
Visits = [120, 135, 150, 160, 155, 170, 165, 180, 190, 200, 210, 205, 220, 230, 240]
```

Task:

Plot a green dotted line (':') with line width = 4.

Add circular markers ('o') of size 10.

Add axis labels and a title "Website Traffic Over 15 Days".

Enable grid lines.




```python
plt.plot(Days, Visits, color='green', linestyle=':', linewidth=4, marker='o', markersize=10)
plt.xlabel("Days")
plt.ylabel("Visits")
plt.grid(True)
plt.title("Website Traffic Over 15 Days")
```

```python
#problem 8: Create a plot comparing Kohli, Rohit Sharma, and Sehwag across 10 years of hypothetical runs.
#Use: Labels Legends Colors Line styles One custom style generate data for this 
import matplotlib.pyplot as plt
import numpy as np
Years = np.arange(2015, 2025)
# Hypothetical runs (random but realistic)
Kohli_runs = [800, 900, 750, 880, 1020, 1100, 970, 1050, 980, 995]
Rohit_runs = [650, 720, 700, 810, 880, 930, 890, 940, 910, 970]
Sehwag_runs = [700, 680, 720, 750, 800, 820, 780, 760, 740, 700]
plt.plot(Years, Kohli_runs, color='red', linestyle='-', marker='^', label='Kohli Runs')
plt.plot(Years, Rohit_runs, color='blue', linestyle='--', marker='s', label='Rohit Runs')
plt.plot(Years, Sehwag_runs, color='green',linestyle=':', marker='o', label='Sehwag Runs')
plt.xlabel("Years")
plt.ylabel("Runs")
plt.legend()
plt.grid(True)
plt.title("Comparing 10 years of runs")
plt.style.use('seaborn-v0_8-poster')# custom style
```
