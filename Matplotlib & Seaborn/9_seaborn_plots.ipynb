#Seaborn comes with built-in example datasets. You can load these datasets directly into a pandas DataFrame using sns.load_dataset().

```python
#How to See All Available Datasets
import seaborn as sns
print(sns.get_dataset_names())
#This will list dataset names like: tips, flights, iris, diamonds, penguins, titanic, etc.
```

```python
#How to Load a Dataset
tips = sns.load_dataset('tips')
tips.head()
#This loads the "tips" dataset (restaurant bills and tips). It returns a pandas DataFrame ready for analysis or plotting.
```

```python
#Now lets look into some plots we can create usign Seaborn
# 1. Line Plot
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title('Line Plot Example')
plt.show()
```

```python
#2. Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time")
plt.title('Scatter Plot with Color by Time')
plt.show()

```

```python
#3. Bar Plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.title('Average Bill per Day')
plt.show()
```

```python
#Box Plot
#Boxplots show distributions, medians, and outliers in one simple plot.

sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Boxplot of Total Bill per Day')
plt.show()
```

```python
#5. Heatmap (Correlation Matrix)
flights = sns.load_dataset('flights')
flights#filghts data
```

```python
pv=flights.pivot(index="month", columns="year", values="passengers")
pv
```

```python
sns.heatmap(pv, annot=True, fmt="d", cmap="YlGnBu")
```

```python
#problem 1: Create a seaborn bar plot showing average marks of students in 4 subjects.
#Add appropriate title and axis labels.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Subjects": ["Math", "Science", "English", "History"],
    "Marks": [85, 78, 92, 74]
})
sns.set_style("whitegrid")
sns.barplot(x="Subjects", y="Marks", edgecolor='black', data=df)
plt.title("Average Marks Of Students in 4 Subjects", fontsize=14)
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
```

```python
#Problem 2: Using seaborn, create a scatter plot of Hours Studied vs Exam Scores.
#Use color "green" and marker size increased.
#Add gridlines and labels.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Hours": [2, 3, 4, 5, 6, 7, 8],
    "Scores": [50, 55, 60, 65, 70, 80, 85]
})
df["size"]=[score*2 for score in df["Scores"]]
sns.scatterplot(x="Hours", y="Scores", color='green', marker='o', size="size", sizes=(50, 300), data=df, legend=False)
plt.title("Hours Studied vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.grid(True)
```

```python
#Problem 3: Create a seaborn line plot of Year vs Sales.Add markers on each point and give the line color "red".
#Add a title "Company Sales Growth".
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Years": [2018, 2019, 2020, 2021, 2022],
    "Sales": [200, 250, 220, 300, 350]
})
sns.lineplot(x="Years", y="Sales", color='red', marker='^', data=df)
plt.title("Company Sales Growth")
```

```python
# Problem 4: Make a seaborn boxplot for Monthly Expenses of 6 People. Show distribution and outliers (if any).
#Add x and y labels.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Expenses": [1200, 1500, 1300, 1700, 1600, 1800]
})
sns.set_style("darkgrid")
sns.boxplot(y="Expenses", data=df)
plt.title("Monthly Expenses of 6 People")
plt.ylabel("Expenses")
```

```python
#Problem 5: Create a seaborn countplot showing the number of students in each class.Plot the count of each class.
#Add a title "Student Distribution by Class"
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Classes": ["Class A", "Class B", "Class A", "Class C", "Class B", "Class A", "Class C", "Class B", "Class A"]
})
sns.countplot(x="Classes", data=df)
plt.title("Student Distribution by Class")
plt.xlabel("Classes")
plt.ylabel("Number of Students")
```
