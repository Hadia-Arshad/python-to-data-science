```python
#basic analysis
```

```python
#total sales per product
# Compute total sales per product
product_sales = df.groupby('Product Category')['Total Amount'].sum()

# Map it back to each row
df['Total_Sales_Per_Product'] = df['Product Category'].map(product_sales)
df
```

```python
customer_counts=df['Customer ID'].value_counts()
# Map to new column
df['Customer_Purchase_Count'] = df['Customer ID'].map(customer_counts)
df
```

```python
#monthly sales
```

```python
# 1. Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
# 2. Create a new 'Month' column (e.g. 2023-05)
df['Month'] = df['Date'].dt.to_period('M')
# 3. Group by 'Month' and calculate total sales
monthly_sales = df.groupby('Month')['Total Amount'].sum()
df['Monthly Sales'] = df['Month'].map(monthly_sales)
df
```

```python
#over all averge sales
df['Average_Sales'] = df['Total Amount'].mean()
df

```

```python
df['Product Sales %'] = (df['Total Amount'] / df['Total Amount'].sum()) * 100
df['Product Sales %'] = df['Product Sales %'].round(2)  # optional: round to 2 decimals
df
```

```python
df.to_csv("updated_sales_retail_dataset.csv")
```
