```python
# Group by product category
product_summary = (
    df.groupby("Product Category")["Total Amount"]
    .sum()
    .reset_index()
    .rename(columns={"Total Amount": "Total Sales"})
)

# Add percentage of total
grand_total = product_summary["Total Sales"].sum()
product_summary["Sales %"] = round(
    (product_summary["Total Sales"] / grand_total) * 100, 2
)

product_summary
```

```python
#Monthly Sales Summary
# Convert Date column
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.to_period("M")

monthly_summary = (
    df.groupby("Month")["Total Amount"]
    .sum()
    .reset_index()
    .rename(columns={"Total Amount": "Total Sales"})
)

monthly_summary
```

```python
#per year sales
# Extract Year from Date
df["Year"] = df["Date"].dt.year

# Group by Year
yearly_summary = (
    df.groupby("Year")["Total Amount"]
    .sum()
    .reset_index()
    .rename(columns={"Total Amount": "Total Sales"})
)

yearly_summary
```

```python
#Customer Purchase Count
customer_summary = (
    df.groupby("Customer ID")["Transaction ID"]
    .count()
    .reset_index()
    .rename(columns={"Transaction ID": "Purchase Count"})
)

customer_summary
```

```python
with pd.ExcelWriter("sales_analysis.xlsx") as writer:
    df.to_excel(writer, sheet_name="Transactions", index=False)
    product_summary.to_excel(writer, sheet_name="Product Summary", index=False)
    monthly_summary.to_excel(writer, sheet_name="Monthly Sales", index=False)
    yearly_summary.to_excel(writer, sheet_name="yearly Sales", index=False)
    customer_summary.to_excel(writer, sheet_name="Customer Summary", index=False)
```
