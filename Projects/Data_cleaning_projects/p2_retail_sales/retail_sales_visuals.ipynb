```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
transaction_df = pd.read_excel("sales_analysis.xlsx", sheet_name="Transactions")
transaction_df
```

```python
product_df = pd.read_excel("sales_analysis.xlsx", sheet_name="Product Summary")
product_df
```

```python
#total sales by product category
plt.figure(figsize=(6,5))
sns.barplot(x="Product Category", y="Total Sales",edgecolor='black',  data=product_df)
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
```

```python
monthly_df = pd.read_excel("sales_analysis.xlsx", sheet_name="Monthly Sales")
monthly_df
```

```python
#Monthly Sales Trend
plt.figure(figsize=(12, 5))
sns.lineplot(x='Month', y='Total Sales', color='red', marker='o', data=monthly_df)
plt.xlabel("Months")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
```

```python
yearly_df = pd.read_excel("sales_analysis.xlsx", sheet_name="yearly Sales")
yearly_df
```

```python
#per year sales 
plt.figure(figsize=(8, 5))
sns.barplot(x='Year', y='Total Sales', edgecolor='black', data=yearly_df)
plt.title("per Year Sales")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
```

```python
#Age Distribution of Customers
transaction_df = pd.read_excel("sales_analysis.xlsx", sheet_name="Transactions")
plt.figure(figsize=(8,5))
sns.histplot(x='Age', bins=20, data=transaction_df)
plt.title("Customer Age Distribution")
plt.xlabel("Customer Age")
plt.ylabel("Frequency")

```
