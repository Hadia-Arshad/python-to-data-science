```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load CSV file into pandas DataFrame
df = pd.read_csv("cleaned_student_results.csv")
df
```

```python
#Bar Chart – Student Average Scores
sns.set_style("whitegrid")
sns.barplot(x='Student', y='Average', edgecolor='black', data=df)
plt.xlabel('Students')
plt.ylabel('Average Score')
plt.title('Students Average Scores')
plt.grid(True, linestyle='--', alpha=0.5)
```

```python
# Horizontal Bar Chart – Student Totals
sns.barplot(y='Student', x='Total', edgecolor='black', data=df)
plt.title('Students Total Marks')
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.grid(True, linestyle='--', alpha=0.6)
```

```python
#pie chart for pass/fail
result_counts = df["Result"].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(result_counts, labels=result_counts.index, autopct="%1.1f%%", 
        startangle=90, colors=["seagreen", "tomato"], explode=[0.05, 0])
plt.title("Pass vs Fail Distribution", fontsize=14)
plt.tight_layout()
```

```python
#stacked Bar Chart – Subject-wise Marks
subjects=['English', 'Math', 'Science']
df.set_index("Student")[subjects].plot(kind="bar", stacked=True, figsize=(10,6), colormap="viridis", edgecolor="black")
plt.title("Subject-wise Scores (Stacked)", fontsize=14)
plt.ylabel("Marks", fontsize=12)
plt.xlabel("Student", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Subjects")
plt.tight_layout()
```
