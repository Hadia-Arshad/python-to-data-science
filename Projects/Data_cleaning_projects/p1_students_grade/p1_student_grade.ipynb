```python
import pandas as pd 
df = pd.read_csv("student_grades.csv")
df
```

```python
df.info() #tells the information about data
```

```python
df = df.dropna().copy() #remove missing data
df
```

```python
df_clean=df.drop_duplicates().copy()
df_clean
```

```python
pivot_df = pd.pivot_table(df, values="Grade", index="Student", columns="Subject", aggfunc="mean")
pivot_df
```

```python
pivot_df = pivot_df.fillna(0)

pivot_df
```

```python
pivot_df["Average"] = pivot_df.mean(axis=1)
pivot_df
```

```python
pivot_df["Result"] = pivot_df["Average"].apply(lambda x: "Pass" if x >= 35 else "Fail")
pivot_df
```

```python
# now lets find Rank for each student 
pivot_df["Rank"] = pivot_df["Average"].rank(ascending=False, method="dense").astype(int)
pivot_df
```

```python
#sort ranking
pivot_df = pivot_df.sort_values("Rank")
pivot_df
```

```python
pivot_df["Total"] = pivot_df[["English","Math","Science"]].sum(axis=1)
pivot_df
```

```python
pivot_df.to_csv("cleaned_student_results.csv")
```
